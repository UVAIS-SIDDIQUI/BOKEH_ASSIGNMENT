```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

ANS= Glyphs in Bokeh are visual representations of data points or shapes that you can add to a plot. They can be circles, squares, lines, wedges, patches, and more, each representing different types of data or relationships within the plotted information.

To add glyphs to a Bokeh plot, you typically create a figure using the figure() function, then call specific glyph methods (like circle(), line(), square(), etc.) on that figure object, providing the necessary data and customization options.

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a new plot with a title and axis labels
plot = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Adding glyphs to the plot
plot.circle(x, y, size=10, color="navy", alpha=0.5, legend_label='Circle')  # Circle glyphs
plot.line(x, y, line_width=2, color="green", legend_label='Line')  # Line glyphs
plot.square(x, y, size=12, color="red", alpha=0.6, legend_label='Square')  # Square glyphs

# Customize the plot attributes
plot.legend.location = "top_left"  # Position of the legend

# Render the plot
show(plot)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

ANS = Customizing the appearance of a Bokeh plot involves modifying various attributes of the plot, such as axes, title, legend, colors, fonts, and more. Here's an example demonstrating how to customize different aspects of a Bokeh plot:

```python
from bokeh.plotting import figure, show
from bokeh.models import Legend

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [4, 5, 3, 2, 6]

# Create a new plot with a title and axis labels
plot = figure(title="Customizing Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

# Customize plot attributes
plot.title.text_color = "blue"  # Title color
plot.title.text_font = "times"  # Title font
plot.title.text_font_size = "20px"  # Title font size

# Customize axes
plot.xaxis.axis_label_text_color = "green"
plot.yaxis.axis_label_text_color = "red"
plot.xaxis.major_label_text_color = "purple"
plot.yaxis.major_label_text_color = "orange"
plot.xaxis.axis_label_text_font = "helvetica"
plot.yaxis.axis_label_text_font = "helvetica"

# Adding glyphs to the plot
plot.line(x, y1, line_width=2, color="blue", legend_label='Line 1')
plot.circle(x, y2, size=10, color="red", alpha=0.5, legend_label='Circle')

# Customize the legend
plot.legend.location = "top_right"
plot.legend.title = "Legend Title"
plot.legend.label_text_font = "times"
plot.legend.label_text_color = "purple"
plot.legend.border_line_color = None
# Render the plot
show(plot)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
# real time?
```

ANS = A Bokeh server is a Python framework that allows you to create interactive web-based visualizations and applications with Bokeh that can be served and updated dynamically. It enables building interactive plots and applications that respond to user interactions or streaming data in real time.

Here's an outline of how you can use a Bokeh server to create interactive plots that can be updated in real time:

Building an Interactive Plot:

 Create a Bokeh plot or application as you would normally, using the Bokeh library in Python.

Define a Bokeh Application:

 To make an application that can be served by the Bokeh server, you need to define a Python script containing the plot or application logic wrapped in a bokeh.application object.

 Start the Bokeh Server:

 Run the Bokeh server and specify the application to serve


 Accessing the Plot/Application:

  Once the server is running, you can access the plot or application by visiting http://localhost:5000 in a web browser (assuming you used port 5000).

Updating the Plot Dynamically:

 With the Bokeh server running, any changes made to the plot logic or data within the defined Bokeh application function (modify_doc in the example) will automatically update the served visualization or application in real time for users accessing it through a web browser.

```python
# Example code snippet to define a Bokeh application
from bokeh.application.handlers import FunctionHandler
from bokeh.application import Application
from bokeh.plotting import figure

def modify_doc(doc):
    # Create or define your Bokeh plot within this function
    plot = figure()
    # ... add glyphs, customize plot, etc.

    doc.add_root(plot)

# Create an application with the modify_doc function as the handler
bokeh_app = Application(FunctionHandler(modify_doc))
```

```python
from bokeh.server.server import Server

server = Server({'/': bokeh_app}, port=5000)  # Map the application to a route
server.start()
```
